```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [2, 3, 4, 5, 6],
        'C': [3, 4, 5, 6, 7],
        'D': [4, 5, 6, 7, 8]}

df = pd.DataFrame(data)

# Create the scatter plot matrix
scatter_matrix(df, diagonal='hist', figsize=(10, 10))

# Display the plot
plt.show()
```
