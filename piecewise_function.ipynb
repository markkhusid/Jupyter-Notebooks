```python
import numpy as np
import matplotlib.pyplot as plt

# Define the piecewise function
def piecewise_function_2(x):
    return np.piecewise(x, 
                        [x < 0, (x >= 0) & (x < 1), x >= 1], 
                        [lambda x: x**2, lambda x: x + 1, lambda x: np.sqrt(x)])

# Define the x-axis values
x = np.linspace(-2, 3, 500)

# Calculate y-axis values using the piecewise function
y = piecewise_function_2(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Piecewise Function')
plt.title('Piecewise Function Example')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
