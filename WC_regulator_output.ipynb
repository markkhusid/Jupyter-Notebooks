```python
import itertools 
  
# initializing list of list  
variables_list = [ ["R1hi", "R1nom", "R1lo"], 
                   ["R2hi", "R2nom", "R2lo"], 
                   ["R3hi", "R3nom", "R3lo"], 
                   ["Iadjhi", "Iadjnom", "Iadjlo"],
                   ["Vrefhi", "Vrefnom", "Vreflo"]
                 ] 
  
# printing lists  
print ("The original lists are : " + str(variables_list)) 
  
# using itertools.product()   
# to compute all possible permutations 
combinations = list(itertools.product(*variables_list)) 
  
# printing result 
counter = 0
for i in combinations:
    print (i)
    counter += 1

print ("There are ", counter, " items.")
```

Assign nominal values to resistors

```python
R1nom = 1000
R2nom = 1000
R3nom = 1000
Iadjnom = 100e-6
Vrefnom = 1.25
```

Assign resistor tolerance (1%)

```python
Rtol = 0.01
Iadjtol = 10e-6
Vreftol = 10e-3
```

Create lists of resistor tolerance bands

```python
R1hi = R1nom*(1+Rtol)
R2hi = R2nom*(1+Rtol)
R3hi = R3nom*(1+Rtol)
Iadjhi = Iadjnom*(1+Iadjtol)
Vrefhi = Vrefnom*(1+Vreftol)
```

```python
R1lo = R1nom*(1-Rtol)
R2lo = R2nom*(1-Rtol)
R3lo = R3nom*(1-Rtol)
Iadjlo = Iadjnom*(1-Iadjtol)
Vreflo = Vrefnom*(1-Vreftol)
```

```python
R1_list = [R1hi, R1nom, R1lo]
R2_list = [R2hi, R2nom, R2lo]
R3_list = [R3hi, R3nom, R3lo]
Iadj_list = [Iadjhi, Iadjnom, Iadjlo]
Vref_list = [Vrefhi, Vrefnom, Vreflo]

print (R1_list)
print (R2_list)
print (R3_list)
print (Iadj_list)
print (Vref_list)
```
