```python
# Python code to generate 5000 Binomial random variables
import numpy as np
import matplotlib.pyplot as plt

p = 0.5  # Probability of success
n = 10 # Number of trials

X = np.random.binomial(n, p, size=5000)  # Generate 5000 Binomial random variables
plt.hist(X, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of Binomial Random Variables')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
# The histogram shows the distribution of the Binomial random variables
# The distribution should resemble a normal distribution as n is large
# The mean of the Binomial distribution is n*p and the variance is n*p*(1-p)
# The mean and variance can be calculated as follows:
mean = n * p
variance = n * p * (1 - p)
print(f'Mean: {mean}')
print(f'Variance: {variance}')
# The mean and variance are printed to the console
# The mean is 5.0 and the variance is 2.5 for n=10 and p=0.5
# The histogram shows the distribution of the Binomial random variables
# The distribution should resemble a normal distribution as n is large
```

```python
# Python code to generate a Binomial PMF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
n = 10  # Number of trials
p = 0.5  # Probability of success

rv = binom(n, p)  # Create a Binomial random variable
x = np.arange(0, n + 1)  # Possible values of the random variable
f = rv.pmf(x)  # PMF of the Binomial random variable
plt.plot(x, f, 'bo', ms=10, label='Binomial PMF')
plt.vlines(x, 0, f, colors='b', lw=5, alpha=0.5)
plt.title('Binomial PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid()
plt.legend()
plt.show()
# The PMF is plotted for the Binomial random variable
# The PMF shows the probability of each possible value of the random variable


```
